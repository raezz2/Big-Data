```python
from pyspark.sql import SparkSession
spark = SparkSession.builder\
        .master("local")\
        .appName("Colab")\
        .config('spark.ui.port', '4050')\
        .getOrCreate()

spark
```

```python
columns = ["language","user_count"]

data = [("java", "20000"), ('pyton', '10000'), ('Scala', '3000')]
rdd = spark.sparkContext.parallelize(data);
```

```python
df = rdd.toDF(columns)
df.printSchema()
df.show()
```

```python
words_list = ['Dog', 'Cat', 'Rabbit', 'Hare', 'Deer', 'Gull', 'Woodpacker','Mole']
words_rdd = spark.sparkContext.parallelize(words_list)
print(words_rdd.count())
```

```python
sc = spark.sparkContext
```

```python
union_inp = sc.parallelize([2,4,5,6,7,8,9])
union_rdd_1 = union_inp.filter(lambda x: x % 2 == 0)
union_rdd_2 = union_inp.filter(lambda x: x % 3 == 0)
print(union_rdd_1.union(union_rdd_2).collect())
```

```python
print(union_rdd_1.union(union_rdd_2).collect())
```

```python
x = sc.parallelize([('a', ['x', 'y', 'z']), ('b',['p','r'])])
x.flatMapValues(lambda y: y).collect()
```

```python
rdd = sc.parallelize(['b', 'a','c'])
sorted(rdd.map(lambda x: (x, 1)).collect())
```

```python
x = sc.parallelize(['a', ['apple', 'banana', 'rrq.lemon']), collect())
```
