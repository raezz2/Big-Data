```python
from pyspark.sql import SparkSession
responsi = SparkSession.builder\
        .master("local")\
        .appName("Responsi")\
        .getOrCreate()
```

```python
sc = responsi.sparkContext
```

```python
flight = responsi.read.csv("/content/flight.csv", header=True)
```

```python
flight.printSchema()
flight.show(20)
```

```python
dfrdd = flight.select("Carrier", "OriginAirportID", "DestAirportID").rdd
```

```python
df = dfrdd.toDF (["Carrier", "OriginAirportID", "DestAirportID"])
df.show()
```
