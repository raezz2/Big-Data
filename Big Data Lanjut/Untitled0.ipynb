```python
from pyspark.sql import SparkSession
spark = SparkSession.builder\
        .master("local")\
        .appName("Colab")\
        .config('spark.ui.port', '4050')\
        .getOrCreate()
```

```python
df = spark.read.csv("/content/FullData.csv", header=True)
df.printSchema()
df.show(5)
```

```python
df.select("Name", "Club").show()
```

```python
rdd = spark.sparkContext.parallelize(df)
df = rdd.toDF()
df.printSchema()
df.show(truncate=False)
```
